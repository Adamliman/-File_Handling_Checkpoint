```
import numpy as np

# 1. Ouvrir le fichier CSV avec le chemin specifique Windows
file_variable = open(r'C:\Users\Liman\Downloads\Loan_prediction_dataset (1).csv', 'r')

try:
    # 2. Charger les donnees numeriques avec genfromtxt
    # skip_header=1 ignore la ligne des titres, usecols=-1 prend la derniere colonne
    loan_amounts = np.genfromtxt(file_variable, delimiter=',', skip_header=1, usecols=-1)

    # 3. Calculs statistiques (nanmean/nanmedian ignorent les cases vides si existantes)
    mean_value = np.nanmean(loan_amounts)
    median_value = np.nanmedian(loan_amounts)
    std_value = np.nanstd(loan_amounts)

    # Affichage des resultats
    print('--- Analyse Statistique du Jeu de Donnees ---')
    print(f'Moyenne (Mean)             : {mean_value:.2f}')
    print(f'Mediane (Median)           : {median_value:.2f}')
    print(f'Ecart-type (Std Deviation) : {std_value:.2f}')

finally:
    # 4. Fermeture obligatoire du fichier
    file_variable.close()
    print('\n[Info] Le fichier a ete ferme avec succes.')
```
